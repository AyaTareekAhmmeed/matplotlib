```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plotting sine curve
x = np.arange(0,3*np.pi,0.1)
y= np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
#plotting cosine
z = np.cos(x)
plt.plot(x,z)
plt.show()
```

```python
#plotting sine&cosine curves
plt.plot(x,y)
plt.plot(x,z)
plt.title('sine and cosine curves')
plt.legend(['Sine','Cosine'])
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
#subplotting sine&cosine curves
plt.subplot(2,1,1)
plt.plot(x,y)
plt.title('sine curve')

plt.subplot(2,1,2)
plt.plot(x,z)
plt.title('cosine curve')

plt.show()
```
